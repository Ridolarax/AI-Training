```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
# Read the Test_file

test_file = pd.read_csv("./45ml_days_test.csv")
```

```python
# Define the training set and the target set
x = test_file.X
y = test_file.Y
```

```python
# Check the overview of the file

test_file.head()
```

```python
# Draw Scatter Plot Using the regplot

sn.regplot(x = "X", y = "Y", data= test_file)
plt.show()
```

```python
# Builds the model

# Initialize the model parameters w and b and assume a learning rate, alpha.

w = 0
b = 0

alpha = 0.001
num_iter = 1000

m = len(x)

for i in range(num_iter):
    func_of_x = w * x + b
    d_dw = (1/m) * sum((func_of_x - y)  * x)
    d_db = (1/m) * sum(func_of_x - y)
    
    # Update the model variables simultaneously
    w = w - (alpha * d_dw)
    b = b - (alpha * d_db)
    
    print(w, ", ", b, ", ", i)
```

```python
def make_prediction(train_data):
    y_values = []
    func_of_x = w * train_data + b
    y_values.append(func_of_x)
    return y_values
```

```python
predictions = make_prediction(x)
predictions = np.array(predictions).T
predictions
```

```python
# Plot the predictions

plt.scatter(x,y)
plt.plot(x, predictions, color = "yellow") 
plt.show()
```
