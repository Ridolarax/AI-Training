```python
# Load the Iris Data

from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("The keys of Iris Dataset: \n{}".format(iris_dataset.keys()))
```

```python
print("Feature names are : \n{}".format(iris_dataset.feature_names))
```

```python
# Print the first five rows of the data

print("The first five datapoints are :\n{}".format(iris_dataset.data[:5]))
```

```python
# Get the shape of the Data

print("The shape of the data is :\n {}".format(iris_dataset.data.shape))
```

```python
# Print out the targets

print(f"The targets are :\n {iris_dataset.target}")
```

```python
# Split the Data to Train and Test Data

from sklearn.model_selection import train_test_split
X = iris_dataset.data
y = iris_dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Check for the shape of the two Train Sets
print("X_train shape is :{}".format(X_train.shape))
print("y_train shape is :{}".format(y_train.shape))
```

```python
# Check for the shape of the two test set

print("X_test shape is :{}".format(X_test.shape))
print("y_test shape is :{}".format(y_test.shape))
```

### Visualization of the Data

```python
# Convert the data to Pandas 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

iris_df = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
iris_df.head()
```

```python
# Plot a ScatterPlot Between Sepal Length and Sepal Width

plt.scatter(x=iris_df["sepal length (cm)"], y = iris_df["sepal width (cm)"])
plt.show()
```

* Since The Scatter Plot doesn't give information about all the features at a go,
* We decide to Use Pairplot

```python
# Create a pairPLot of the DataFrame

sns.pairplot(iris_df, height= 3)
plt.show()
```

```python
# Import the Knn Neighbor classifier class 

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# Train the classifier Class

knn.fit(X_train, y_train)
```

```python
# Test a new dataset
X_new = np.array([[5, 2.9, 1, 0]])
print("The shape is of the new instance is:\n{}".format(X_new.shape))
```

```python
# Make the prediction

prediction = knn.predict(X_new)
print(prediction)
```

```python
# Determine what the prediction relates to:

print("The model has predicted the specie to be: \n{}".format(iris_dataset["target_names"][prediction]))
```

### Evaluating the model

```python
# Evaluate the correctness of the model

print("The accuracy of the model is {}".format(knn.score(X_test, y_test)))
```
